     - Year: 17, integer
     - Player: 123, object (string)
     - Pos: 11, object (string)
     - HT: 18, object (number separated by -)
     - WT: 49, integer

# What is the average weight of all players who played during this time?

### Average weight of ALL players

```python
all_avg_weight = nba_data["WT"].mean()
print(f"Average of ALL weights of ALL players during the entire 2000 - 2016 time period is {round(all_avg_weight)} lbs.")
```

### Average weight of players each year

```python
avg_weight_per_year = nba_data.groupby('Year', as_index = False).mean().rename(columns = {"WT": "Average Weight"})
avg_weight_per_year
```

### Average weight of each player who played during this time

```python
pd.set_option('display.max_rows', None)

avg_weight = nba_data.groupby('Player', as_index = False).mean().rename(columns = {"WT": "Average Weight"})
avg_weight[['Player','Average Weight']]
```

### double checking the calculation above

```python
stephen_curry = nba_data[nba_data['Player'] == 'Stephen Curry']
print(stephen_curry[["Year","Player", "WT"]])

stephen_curry_avg_weight = avg_weight[avg_weight['Player'] == 'Stephen Curry']
print(stephen_curry_avg_weight)
```

# What team is represented the most during this time?

### Miami Heat is represented the most, followed by Boston Celtics, San Antonio Spurs and Los Angeles Lakers.

```python
team_count = nba_data.groupby(["Team"], as_index = False).count().rename(columns = {"Player": "Number of Players"})
team_count[["Team", "Number of Players"]].sort_values(by = "Number of Players", ascending = False).head(5)
```

# What draft class is represented the most during this time?

### 1996 Rnd 1 Pick 13 is represented the most!

```python
draft_class = nba_data.groupby(["NBA Draft Status"], as_index = False).count().rename(columns = {"Player": "Number of Players"})
draft_class[["NBA Draft Status", "Number of Players"]].sort_values(by = "Number of Players", ascending = False).head(5)
```

# Are foreign players more prevalent in the All-Star Game during this time?

```python
nationalities = nba_data["Nationality"].unique()
print(nationalities)
```

### For this analysis, I defined foreign players as those without United States citizenship/nationality.

### There are more players with US Citizenship in the All-Star Game during this time.

```python
nba_data ["Citizenship"] = ["US" if "United States" in x else "Foreign" for x in nba_data["Nationality"]]
us_citizenship = nba_data.groupby("Citizenship", as_index = False).count().rename(columns = {"Player":"Number of Players"})
us_citizenship[["Citizenship", "Number of Players"]]
```

# How often are the Western All-Stars voted in by fans?

### 39% of the time

```python
selection = nba_data.groupby("Selection Type", as_index = False).count().rename(columns = {"Player":"Number of Players"})
selection = selection[["Selection Type", "Number of Players"]]
print(selection)
```

```python
western_all_star = selection["Number of Players"][3:].sum()
western_fans = selection["Number of Players"][4]

print(western_fans)
print(western_all_star)

print(f'Western All-Stars were voted in by fans {round(100 * western_fans/western_all_star)}% of the time.')
```
